## Question 4: Create a loyalty flag

### Use the Transform Function

```python
# Use the transform function to create a new column
ords_prods_merge['max_order'] = ords_prods_merge.groupby(['user_id'])['order_number'].transform(np.max)
```

```python
# Show results of the transform function
ords_prods_merge.head(60)
```

## Use the Loc Function

```python
# Flag customers with over 40 orders as 'Loyal customer' 
ords_prods_merge.loc[ords_prods_merge['max_order'] > 40, 'loyalty_flag'] = 'Loyal customer' 
```

```python
# Flag customers with more than 10 and up to 40 orders as 'Regular customer' 
ords_prods_merge.loc[(ords_prods_merge['max_order'] <= 40) & (ords_prods_merge['max_order'] > 10), 'loyalty_flag'] = 'Regular customer' 
```

```python
# Flag customers with 10 or fewer orders as 'New customer' 
ords_prods_merge.loc[ords_prods_merge['max_order'] <= 10, 'loyalty_flag'] = 'New customer' 
```

```python
# Check value counts of loyalty flag 
ords_prods_merge['loyalty_flag'].value_counts(dropna = False) 
```

```python
# Check head of certain columns
ords_prods_merge[['user_id', 'loyalty_flag', 'order_number']].head(60)
```

## Question 6: Create a spending flag

```python
# Use the transform function to create a new column of mean prices
ords_prods_merge['mean_order'] = ords_prods_merge.groupby(['user_id'])['prices'].transform(np.mean)
```

```python
# Show head
ords_prods_merge.head()
```

```python
# change mean_order column name to average_prices
ords_prods_merge.rename(columns={'mean_order': 'average_prices'}, inplace = True)
```

```python
# Flag customers with a mean product price of under 10 as 'Low spender' 
ords_prods_merge.loc[ords_prods_merge['average_prices'] < 10, 'spending_flag'] = 'Low spender'
```

```python
# Flag customers with a mean product price of 10 or more as 'High spender' 
ords_prods_merge.loc[ords_prods_merge['average_prices'] >= 10, 'spending_flag'] = 'High spender'
```

```python
# Show value counts of high and low spenders
ords_prods_merge['spending_flag'].value_counts(dropna = False)
```

```python
# Show head
ords_prods_merge.head()
```

## Question 7: Create an order frequency flag

```python
# Use the transform function to create a new column of median days since prior order
ords_prods_merge['median_days_since_prior_order'] = ords_prods_merge.groupby(['user_id'])['days_since_prior_order'].transform(np.median)
```

```python
# show head
ords_prods_merge.head()
```

```python
# Flag customers with a median of over 20 days since their prior order as 'Non-frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] > 20, 'frequency_flag'] = 'Non-frequent customer'
```

```python
# Flag customers with a median of over 10 and 20 or fewer days since their prior order as 'Regular customer'
ords_prods_merge.loc[(ords_prods_merge['median_days_since_prior_order'] > 10) & (ords_prods_merge['median_days_since_prior_order'] <= 20), 'frequency_flag'] = 'Regular customer'
```

```python
# Flag customers with a median of 10 or fewer days since their prior order as 'Frequent customer'
ords_prods_merge.loc[ords_prods_merge['median_days_since_prior_order'] <= 10, 'frequency_flag'] = 'Frequent customer'
```

```python
# Check value counts
ords_prods_merge['frequency_flag'].value_counts(dropna = False)
```
