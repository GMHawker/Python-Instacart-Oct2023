## Question 5:

### Create a sample

```python
# Create a seed

np.random.seed(4)
```

```python
# create a  list holding true/false values

dev = np.random.rand(len(df)) <= 0.7
```

```python
# Store 70% of the sample in the dataframe big

big = df[dev]
```

```python
#  Store 30% of the sample in the dataframe small

small = df[~dev]
```

```python
# Check full database shape

len(df)
```

```python
# Check shape of samples

len(big) + len(small)
```

```python
# Create a sample subset as df_sub

df_sub = small[['order_hour_of_day', 'prices']]
```

```python
# Check data types

df_sub.dtypes
```

```python
# Create a line chart from sample

line = sns.lineplot(data = df_sub, x = 'order_hour_of_day',y = 'prices')
```
