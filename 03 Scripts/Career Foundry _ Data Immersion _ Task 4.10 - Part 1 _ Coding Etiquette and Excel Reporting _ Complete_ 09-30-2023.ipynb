### Eplaining qualifier names:

#### Young: Age < 40
#### Older: Age >= 40
#### Dependant: family_status is 'living with parents and siblings'
#### Single: family_status is 'single' or 'divorced/widowed'
#### Married: family_status is 'married'
#### Childless: dependants = 0
#### Parent: dependants >= 1

```python
# Flag everyone under 40 as young

df_profile.loc[df_profile['age'] < 40, 'age_group'] = 'young' 
```

```python
# Flag everyone over 40 as older

df_profile.loc[df_profile['age'] >= 40, 'age_group'] = 'older' 
```

```python
# Flag eveyone living with parents and siblings as a dependant

df_profile.loc[df_profile['family_status'].isin
               (['living with parents and siblings']),
               'family_role'] = 'dependant'
```

```python
# Flag everyone married as married

df_profile.loc[df_profile['family_status'].isin
               (['married']),
               'family_role'] = 'married'
```

```python
# Flag everyone 'single'  or 'divorced/widowed' as single

df_profile.loc[df_profile['family_status'].isin
               (['single', 'divorced/widowed']),
               'family_role'] = 'single'
```

```python
# Flag everyone with 0 dependants as childless

df_profile.loc[df_profile['dependants'] == 0, 'parental_status'] = 'childless' 
```

```python
# Flag everyone with >= 1 dependants as parent

df_profile.loc[df_profile['dependants'] >= 1, 'parental_status'] = 'parent' 
```
