### If-Statement

```python
# Create an if-statement to identify product price range
def price_label(row):

  if row['prices'] <= 5:
    return 'Low-range product'
  elif (row['prices'] > 5) and (row['prices'] <= 15):
    return 'Mid-range product'
  elif row['prices'] > 15:
    return 'High range'
  else: return 'Not enough data'
```

```python
# Apply if-statement to subset
df['price_range'] = df.apply(price_label, axis = 1)
```

```python
# Check price range value counts within the subset
df['price_range'].value_counts(dropna = False)
```

```python
# Check the maximum price within the subset
df['prices'].max()
```

### Loc() Statement

```python
# Create a loc() statement for High range products
df.loc[df['prices'] > 15, 'price_range_loc'] = 'High-range product'
```

```python
# Create a loc() statement for Mid range products
df.loc[(df['prices'] <= 15) & (df['prices'] > 5), 'price_range_loc'] = 'Mid-range product' 
```

```python
# Create a loc() statement for Low range products
df.loc[df['prices'] <= 5, 'price_range_loc'] = 'Low-range product'
```

```python
# Check the loc() price range value counts within the subset
df['price_range_loc'].value_counts(dropna = False)
```

## Question 4: Create new column called "busiest_period_of_day"

```python
# Create a for-loop to show the top 3 and bottom 3 busiest hours
result = []

for value in ords_prods_merge['order_hour_of_day']:
  if value == 10:
    result.append('Most orders')
  elif value == 11:
    result.append('Most orders')
  elif value == 14:
    result.append('Most orders')
  elif value == 3:
    result.append('Fewest orders')
  elif value == 4:
    result.append('Fewest orders')
  elif value == 2:
    result.append('Fewest orders')
  else:
    result.append('Average Orders')
```
